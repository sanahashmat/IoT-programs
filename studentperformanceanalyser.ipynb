```python
# student performance analyser

students = [
    {"name": "Arun", "math": 85, "vlsi": 78, "python": 91},
    {"name": "Vijay", "math": 72, "vlsi": 88, "python": 69},
    {"name": "Kiran", "math": 90, "vlsi": 95, "python": 89},
    {"name": "Rahul", "math": 60, "vlsi": 65, "python": 70}
]

def calculate_total(student):
    return student["math"] + student["vlsi"] + student["python"]

for s in students:
    s["total"] = calculate_total(s)

students_sorted = sorted(students, key=lambda x: x["total"], reverse=True)

topper = students_sorted[0]
lowest = students_sorted[-1]

subject_highest = {
    "math": max(students, key=lambda s: s["math"]),
    "vlsi": max(students, key=lambda s: s["vlsi"]),
    "python": max(students, key=lambda s: s["python"])
}

with open("student_report.txt", "w") as f:
    f.write("=== Student Performance Report ===\n\n")
    f.write("Sorted by Total Marks:\n")
    for s in students_sorted:
        f.write(f"{s['name']} -> {s['total']}\n")

    f.write("\nTopper: " + topper["name"] + "\n")
    f.write("Lowest Performer: " + lowest["name"] + "\n\n")

    f.write("Subject-wise Highest:\n")
    for subject, student in subject_highest.items():
        f.write(f"{subject} : {student['name']} ({student[subject]})\n")

print("student_report.txt generated.")
```
