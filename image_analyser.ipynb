```python
import os
import csv
from PIL import Image
import matplotlib.pyplot as plt

# Your image folder path
folder_path = "/content/"
image_info = []

# 1. Read all .jpg and .png files
for filename in os.listdir(folder_path):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        img_path = os.path.join(folder_path, filename)

        # Open image
        img = Image.open(img_path)
        width, height = img.size

        # 2. Display image
        plt.imshow(img)
        plt.title(filename)
        plt.axis("off")
        plt.show()

        # 3. Store details in dictionary
        image_info.append({
            "filename": filename,
            "width": width,
            "height": height,
            "area": width * height
        })

# 4. Largest & smallest image
largest_image = max(image_info, key=lambda x: x["area"])
smallest_image = min(image_info, key=lambda x: x["area"])

print("\nLargest Image:", largest_image)
print("Smallest Image:", smallest_image)

# 5. Save extracted information into CSV
csv_path = os.path.join(folder_path, "image_sizes.csv")

with open(csv_path, "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["filename", "width", "height", "area"])
    writer.writeheader()
    writer.writerows(image_info)

print("\nCSV file saved at:", csv_path)
```
